## Association Rules Method for Books Dataset

```python
!pip install mlxtend
```

```python
#importing basic libraries and functions 
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#importing and analysing data 
data = pd.read_csv("book.csv")
data
```

```python
data.info()
```

```python
# Apriori Algorithm
frequent_itemsets = apriori(data, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
# Level of Confidence and lift
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.2)
rules.head()
```

```python
rules.sort_values("lift", ascending=False).head()
```

```python
rules[["support","confidence","lift"]].hist()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('classic')
sns.scatterplot(data=rules,x='support',y='confidence',hue='lift', s= 100)
plt.show()
```

## changing minimum length of apriori algorithum  and applying visualizations

```python
# Level of Confidence and lift :
rules= association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)[0:20]
rules
```

```python
rules[rules.lift>1]
```

```python
(list(range(1,11)),frequent_itemsets.itemsets[1:11])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(x=list(range(1,11)),height=frequent_itemsets.support[1:11]);plt.xticks
```

# <font color= "voilet"> Association rules for Movies data set

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data1 = pd.read_csv("my_movies.csv")
data1
```

```python
data1.info()
```

```python
data2 = data1.iloc[:,5:]
data2
```

```python
# Apriori Algorithm
frequent_itemsets = apriori(data2, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
# Level of Confidence and Lif
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.2)
rules
```

```python
rules.sort_values("lift", ascending=False).head()
```

```python
#plotiing histogram 
rules[['support','confidence','lift']].hist()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('classic')
sns.scatterplot(data=rules,x='support',y='confidence',hue='lift', s= 100)
plt.show()
```
